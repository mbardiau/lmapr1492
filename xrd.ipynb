```python
from pymatgen import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import *
```

```python
#téléchargement de la structure
structure = MPRester("KGkLFN8rUMTD0Z1V").get_structure_by_material_id("mp-20289")
file = CifWriter(structure,0.1)
file.write_file("download.cif")
```

```python
#Initialisation des variables utiles
ray = XRDCalculator(wavelength = "CuKa",symprec=1)
pattern = ray.get_pattern(structure)
indices = pattern.hkls
```

Le système étudié étant un système rhombohédrique, la notation de Bravais-Miller (h k i l) est utilisée. Celle-ci comporte 4 indices au lieu de trois, où i est un indice redondant qui obéit à la relation h + k + i = 0. 

```python
#Affichage des caractéristiques des pics
print("Premier pic :\n\n 2θ = {}°\nIntensité = {}CPS\nIndices de Bravais-Miller : {}\n Multiplicité : {}\n".format(pattern.x[0], pattern.y[0], indices[0][0]["hkl"], indices[0][0]["multiplicity"]))

print("Deuxième pic :\n\n 2θ = {}°\nIntensité = {}CPS\nIndices de Bravais-Miller : {}\n Multiplicité : {}\n".format(pattern.x[1], pattern.y[1], indices[1][0]["hkl"], indices[1][0]["multiplicity"]))

print("Troisième pic :\n\n 2θ = {}°\nIntensité = {}CPS\nIndices de Bravais-Miller : {}\n Multiplicité : {}\n".format(pattern.x[2], pattern.y[2], indices[2][0]["hkl"], indices[2][0]["multiplicity"]))
```

Note : l'intensité est donnée en CPS (counts per second).

```python
#Affichage du plot
ray.show_plot(structure, two_theta_range = [0,28]) #Limité à 2theta = 28° afin de voir uniquement les 3 premiers pics
```
